```python
df = pd.read_csv('../input/train_relationships.csv')
```

```python
df.head()
```

```python
cols = ['family', 'p1', 'p2'] # new DF with family index
new_df = pd.DataFrame(columns=cols)
for i in range(len(df)):
    p1 = df.iloc[i].p1.split('/')
    p2 = df.iloc[i].p2.split('/')
    new_df.loc[i] = [p1[0], p1[1], p2[1]] # family, first person, kinship relation person
```

```python
print('Number of families: {}'.format(new_df.family.nunique()))
```

```python
f = new_df.family.unique()
number_of_relations = []
for family in f:
    number_of_relations.append(new_df.loc[new_df.family == family].values.shape[0])
plt.hist(number_of_relations, bins=100) # most families have 1-10 relations, there are entries for 70+ relations!
print("checking number of entries: ", len(df), np.cumsum(number_of_relations)[-1])
plt.show()
```
